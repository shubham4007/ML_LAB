```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset = pd.read_csv("zoo.csv")
dataset.head()
```

```python
dataset.shape
```

```python
classes = ['Mammal', 'Bird', 'Reptile', 'Fish', 'Amphibian', 'Bug', 'Invertebrate']
```

```python
dataset.describe()
```

```python
from sklearn.model_selection import train_test_split
```

```python
print(dataset)
```

```python
X= dataset.drop('class_type', axis=1).drop('animal_name', axis=1)
```

```python
dataset['class_type']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, 
    dataset['class_type'], 
    test_size=0.15, random_state=42
)
```

```python
X_train.head()
```

```python
from sklearn import tree
```

```python
classifier = tree.DecisionTreeClassifier()
```

```python
classifier.fit(X_train, y_train)

tree.plot_tree(classifier)
```

```python
import matplotlib.pyplot as plt

fig=plt.figure(figsize=(25,20))
tree.plot_tree(classifier,feature_names=X_train.columns,class_names=classes,filled=True)
```

```python
y_pred = classifier.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
print("Actual class \t\t\t\t Predicted class")

y_test = list(y_test)

for i in range(len(y_test)):
    print(f"{classes[y_test[i]-1]}\t\t\t\t\t{classes[y_pred[i]-1]}")
```
